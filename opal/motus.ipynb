```python
import pandas as pd
```

```python
motus=pd.read_csv('/shared/projects/mudis4ls_is4_benchmark/test_results/motus/results.txt', sep='\t')
```

```python
motus_species = motus[motus["#mOTUs2_clade"].str.contains(r"s__.+$", na=False)]
```

```python
motus_species=motus_species.replace(r"k__", "d__", regex=True)
```

```python
motus_species.rename(columns={"#mOTUs2_clade": "gtdb_r220_classification","ERS12377136":"abundance"}, inplace=True)
```

```python
motus_species['species'] = motus_species['gtdb_r220_classification'].str.extract(r's__([^;]+)')
```

```python
motus_species['species'] = motus_species['species'].str.replace(r'\s*\[[^\]]*\]$', '', regex=True)
```
