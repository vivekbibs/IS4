```python
# Créer un dictionnaire de conversion
mapping_dict = pd.Series(gtdb_mapping["taxonomy_220"].values, index=gtdb_mapping["taxonomy_207"]).to_dict()
```

```python
mapping_dict
```

```python
mpa_output["#clade_name"] = mpa_output["#clade_name"].map(mapping_dict).fillna(mpa_output["#clade_name"])
```

```python
mpa_output
```

```python
# 🔹 Extraire des mappings pour les niveaux taxonomiques supérieurs (Phylum, Classe, etc.)
mapping_levels = {}
for old_tax, new_tax in mapping_dict.items():
    old_parts = old_tax.split(";")
    new_parts = new_tax.split(";")
    for i in range(len(old_parts)):
        old_sub = ";".join(old_parts[:i+1])
        new_sub = ";".join(new_parts[:i+1])
        if old_sub not in mapping_levels:
            mapping_levels[old_sub] = new_sub

# 🔥 Fonction pour mapper les taxonomies partiellement
def update_taxonomy(tax):
    if tax in mapping_dict:
        return mapping_dict[tax]  # Correspondance complète
    for i in range(len(tax.split(";")), 0, -1):  # Vérifier du plus long au plus court
        sub_tax = ";".join(tax.split(";")[:i])
        if sub_tax in mapping_levels:
            return tax.replace(sub_tax, mapping_levels[sub_tax], 1)
    return tax  # Pas de correspondance trouvée

# Appliquer la fonction au dataframe
mpa_output["#clade_name"] = mpa_output["#clade_name"].apply(update_taxonomy)
```

```python
#Je veux mettre le mapping complet dans un fichier tsv
# car pour l'instant dans le tsv ya que les chemins jusqu'a espèce complet
# ya pas le mapping entre juste firmicutes et juste bacillota
#





# Stocker les mappings pour tous les niveaux taxonomiques
mapping_levels = {}

for old_tax, new_tax in mapping_dict.items():
    old_parts = old_tax.split(";")
    new_parts = new_tax.split(";")
    
    for i in range(len(old_parts)):
        old_sub = ";".join(old_parts[:i+1])
        new_sub = ";".join(new_parts[:i+1])
        mapping_levels[old_sub] = new_sub

# Créer une liste pour stocker les correspondances
mapping_data = []

# Générer les correspondances pour chaque niveau taxonomique
for old_tax, new_tax in mapping_levels.items():
    level = old_tax.count(";") + 1  # Déterminer le niveau taxonomique (ex: 1 = domaine, 7 = espèce)
    mapping_data.append([old_tax, new_tax, level])

# Convertir en DataFrame
mapping_df = pd.DataFrame(mapping_data, columns=["old_taxonomy", "new_taxonomy", "level"])

# Sauvegarder le mapping sous forme de fichier TSV
mapping_df.to_csv("/shared/projects/mudis4ls_is4_benchmark/gtdb/207_220_mapping_full.tsv", sep="\t", index=False)

print("✅ Fichier 'taxonomy_mapping.tsv' créé avec succès !")
```
