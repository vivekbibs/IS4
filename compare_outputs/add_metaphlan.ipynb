```python
import pandas as pd
```

```python
metaphlan=pd.read_csv("/shared/projects/mudis4ls_is4_benchmark/test_results/metaphlan/metaphlan_output_gtdb.txt", sep="\t",header=1)
```

```python
metaphlan.rename(columns={'#clade_name':'gtdb_r220_classification'}, inplace=True)
```

```python
metaphlan.rename(columns={'relative_abundance':'abundance_metaphlan'}, inplace=True)
```

```python
meteor_sylph_bracken=pd.read_csv(
    "/shared/projects/mudis4ls_is4_benchmark/test_results/compare_outputs/sylph_meteor_bracken/sylph_meteor_bracken.tsv",
    sep="\t"
    )
```

```python
meteor_sylph_bracken
```

```python
all=meteor_sylph_bracken.merge(metaphlan, on="gtdb_r220_classification", how="outer")
```

```python
all
```

```python
cols = list(all.columns)

# Déplacer les colonnes 'x' et 'y' en 2e et 3e position
new_order = ["species","abundance_metaphlan"] + [col for col in cols if col not in ["species","abundance_metaphlan"]]
# new_order = ["species"] + [col for col in cols if col not in ["species"]]

# Réorganiser le DataFrame
all = all[new_order]
```

```python
all
```

```python
import matplotlib.pyplot as plt
from matplotlib_venn import venn3, venn3_circles  # Pour 3 ensembles
import matplotlib.patches as patches
import pandas as pd

# Supposons que ton DataFrame s'appelle df

# Définir les ensembles en fonction des colonnes non-NaN
kraken_set = set(all[all["abundance_kraken"].notna()].index)
sylph_set = set(all[all["abundance_sylph"].notna()].index)
meteor_set = set(all[all["abundance_meteor"].notna()].index)
metaphlan_set = set(all[all["abundance_metaphlan"].notna()].index)

# Création de la figure
fig, ax = plt.subplots(figsize=(8, 8))

# Ajout des ellipses pour simuler un Venn à 4 ensembles
colors = ["red", "blue", "green", "purple"]
labels = ["Kraken", "Sylph", "Meteor", "Metaphlan"]

ellipses = [
    patches.Ellipse((0.3, 0.6), width=0.7, height=0.4, angle=30, color=colors[0], alpha=0.5, label=labels[0]),
    patches.Ellipse((0.7, 0.6), width=0.7, height=0.4, angle=-30, color=colors[1], alpha=0.5, label=labels[1]),
    patches.Ellipse((0.4, 0.4), width=0.7, height=0.4, angle=-30, color=colors[2], alpha=0.5, label=labels[2]),
    patches.Ellipse((0.6, 0.4), width=0.7, height=0.4, angle=30, color=colors[3], alpha=0.5, label=labels[3])
]

for e in ellipses:
    ax.add_patch(e)

# Ajout des annotations (Nombre d'assignations taxonomiques en commun)
ax.text(0.5, 0.75, len(kraken_set & sylph_set), ha="center", fontsize=12, color="black")
ax.text(0.5, 0.25, len(meteor_set & metaphlan_set), ha="center", fontsize=12, color="black")
ax.text(0.2, 0.5, len(kraken_set & meteor_set), ha="center", fontsize=12, color="black")
ax.text(0.8, 0.5, len(sylph_set & metaphlan_set), ha="center", fontsize=12, color="black")
ax.text(0.5, 0.5, len(kraken_set & sylph_set & meteor_set & metaphlan_set), ha="center", fontsize=14, fontweight="bold", color="black")

# Personnalisation et légende
ax.set_xlim(0, 1)
ax.set_ylim(0, 1)
ax.set_xticks([])
ax.set_yticks([])
ax.set_frame_on(False)
ax.legend(loc="upper right")

# Affichage du diagramme
plt.title("Diagramme de Venn des Assignations Taxonomiques (Kraken, Sylph, Meteor, Metaphlan)")
plt.show()
```
